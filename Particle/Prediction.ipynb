```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import Pred as pr
```

```python
with open('Results_Burgers.pkl','rb') as f:  # Python 3: open(..., 'rb')
    theta1, theta2, save,res,sig,n,m,thetavar,sigma,t = pickle.load(f)
```

```python
N_CG=128 # Number of CG states (64 for AD)
```

```python
X_axis=np.linspace(-1+1/N_CG,1-1/N_CG,N_CG)
```

```python
#Do Prediction
mean_pred,std_pred=pr.prediction(n[0,5,:],sig,theta1,theta2,thetavar,25)
```

```python
#Visualize Results
i=23 #Choose prediction time i+1 time steps into the future
plt.rc('font', family='serif')
plt.plot(X_axis,m[i+1,5,:]/500000,label='Reference')
plt.plot(X_axis,mean_pred[i,:],label='Predicted')
plt.fill_between(X_axis,(mean_pred[i,:]-3*std_pred[i,:]).reshape((128,)),(mean_pred[i,:]+3*std_pred[i,:]).reshape((128,)),alpha=0.3, edgecolor='#CC4F1B', facecolor='#CC4F1B',linewidth=0,label='+/- 3 Standard Deviation')
plt.ylabel('X')
plt.xlim(-1,1)
plt.legend(loc="best")
```
